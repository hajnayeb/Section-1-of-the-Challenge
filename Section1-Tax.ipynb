```python
import pandas as pd
df=pd.read_csv('https://www.irs.gov/pub/irs-soi/16zpallagi.csv')
```

```python
pd.set_option('display.max_columns', 0)
pd.set_option('display.max_rows', 15)
display(df[1:8])
```

```python
df=df.loc[(df['zipcode']!=0),['STATEFIPS','STATE','zipcode','agi_stub','N1','A02650','A06500']]
pd.set_option('display.max_columns', 0)
pd.set_option('display.max_rows', 10)

display(df)
```

```python
stgroups = df.groupby(by='STATEFIPS')
w = df.groupby('STATE')['A02650'].sum()
n = df.groupby('STATE')['N1'].sum()
x=w.index
y=w.values
taxpayers=n.values
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.subplot(211)
plt.bar(x, y)
plt.xlabel('State Name'); plt.ylabel('Total Income [USD]')
plt.subplot(212)
plt.bar(x, taxpayers)
plt.xlabel('State Name'); plt.ylabel('Total Number of Taxpayers')
plt.show()
```
